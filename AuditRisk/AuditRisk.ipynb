### However, The risk column in both dataset contains different vaues. So, we rename risk column of trial dataset as Risk_Trial to avoid conflicting issue while merging

```python
# Rename trial dataset columns
trial_df.columns = ['Sector_score','LOCATION_ID', 'PARA_A', 'Score_A', 'PARA_B',
       'Score_B',  'TOTAL', 'numbers', 'Marks',
       'Money_Value', 'MONEY_Marks', 'District',
       'Loss', 'LOSS_SCORE', 'History', 'History_score', 'Score', 'Risk_trial']
```

```python
trial_df.head(10)
```

### As we can see SCORE_A & SCORE_B from trial contains values that are multiples of 10 comparing to audit_risk

```python
# Update SCORE_A & SCORE_B column from trial
trial_df['Score_A'] = trial_df['Score_A']/10
trial_df['Score_B'] = trial_df['Score_B']/10
```

```python
trial_df.head(20)
```

```python
trial_df.describe()
```

```python
audit_df.describe()
```

```python
# Finding common columns of trial & audit_risk that can be merged
common_columns = np.intersect1d(audit_df.columns, trial_df.columns)
common_columns
```

### As we can see the common columns contains the same values from the descriptions so we can merge them now

```python
# Merge trial & audit_risk Dataframes  on common columns  using outer join
merged = pd.merge(audit_df, trial_df, how='outer', on = ['History', 'LOCATION_ID', 'Money_Value', 'PARA_A', 'PARA_B',
       'Score', 'Score_A', 'Score_B', 'Sector_score', 'TOTAL', 'numbers'])
merged.columns
```

```python
merged.info()
```

```python
# Checking missing value of Money_Value column
merged[merged['Money_Value'].isnull() == True]
```

```python
#Replacing the missing value by the median of the Money_Value column
merged['Money_Value'] = merged['Money_Value'].fillna(merged['Money_Value'].median())
```

```python
merged.info()
```

```python
# Check if there any column left contains missing value
merged.isnull().sum()
```

```python
# Unique values that each column contains
merged.nunique()
```

### Here detection_risk contains only a single value so we can discard this column from the dataframe

```python
# drop Detection_Risk column
merged_df = merged.drop(['Detection_Risk'], axis = 1)
```

```python
merged_df.info()
```

### Here Location_ID is object tyoe. So we need to fix that

```python
#Unique values in LOCATION_ID column
merged_df["LOCATION_ID"].unique()
```

### So location_id contains both numeric & non numeric values

```python
# Total Number of Non numeric values
print(len(merged_df[(merged_df["LOCATION_ID"] == 'LOHARU') | (merged_df["LOCATION_ID"] ==  'NUH') | (merged_df["LOCATION_ID"] == 'SAFIDON')]))
```

### As the Location_id column contains only 3 non numeric value so deleting the rows containing non numeric locaton_id would be a better idea as there is no way to sequence them in order

```python
# Delete rows with non numeric location_id
merged_df = merged_df[(merged_df.LOCATION_ID != 'LOHARU')]
merged_df = merged_df[(merged_df.LOCATION_ID != 'NUH')]
merged_df = merged_df[(merged_df.LOCATION_ID != 'SAFIDON')]
merged_df = merged_df.astype(float)
```

```python
merged_df.info()
```

### As we merged audit_risk & trial dataframe so there is a possibility of occurance of duplicate values. In such case we need to drop the duplicate rows.

```python
# Dropping duplicates
merged_df = merged_df.drop_duplicates(keep = 'first')
```
