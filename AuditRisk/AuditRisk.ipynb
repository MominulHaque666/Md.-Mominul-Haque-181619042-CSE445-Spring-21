### Method to findout correlative independent features 

```python
def correlation(dataset, threshold):
    col_corr = []
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) >= threshold:
                colname = corr_matrix.columns[i]
                col_corr.append(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.85)
len(set(corr_features))
```

### Alternative Method to find correlative features

```python
# Find correlated features
def correlated_features(dataset, threshold, target_variable):
    df = dataset.corr().loc[target_variable].drop(target_variable)
    return df[df.abs() >= threshold]
```
