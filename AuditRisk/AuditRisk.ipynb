### Here comparing the description of audit_risk & trial dataset, there are several columns that are matching & can be merged as well. To merge them, column names of trial dataset should be modified first with respect to Audit_risk dataset

```python
# Rename trial dataset columns
trial_df.columns = ['Sector_score','LOCATION_ID', 'PARA_A', 'Score_A', 'PARA_B',
       'Score_B',  'TOTAL', 'numbers', 'Marks',
       'Money_Value', 'MONEY_Marks', 'District',
       'Loss', 'LOSS_SCORE', 'History', 'History_score', 'Score', 'Risk_trial']
```

```python
trial_df.head(10)
```

### As we can see SCORE_A & SCORE_B from trial contains values that are multiples of 10 comparing to audit_risk

```python
# Update SCORE_A & SCORE_B column from trial
trial_df['Score_A'] = trial_df['Score_A']/10
trial_df['Score_B'] = trial_df['Score_B']/10
```

```python
trial_df.head(20)
```

```python
trial_df.describe()
```

```python
audit_df.describe()
```

```python
# Finding common columns of trial & audit_risk that can be merged
common_columns = np.intersect1d(audit_df.columns, trial_df.columns)
common_columns
```

### As we can see the common columns contains the same values from the descriptions so we can merge them now

```python
# Merge trial & audit_risk Dataframes  on common columns  using outer join
merged = pd.merge(audit_df, trial_df, how='outer', on = ['History', 'LOCATION_ID', 'Money_Value', 'PARA_A', 'PARA_B',
       'Score', 'Score_A', 'Score_B', 'Sector_score', 'TOTAL', 'numbers'])
merged.columns
```

```python
merged.info()
```

```python
# Checking missing value of Money_Value column
merged[merged['Money_Value'].isnull() == True]
```

```python
#Replacing the missing value by the median of the Money_Value column
merged['Money_Value'] = merged['Money_Value'].fillna(merged['Money_Value'].median())
```

```python
merged.info()
```

```python
# Check if there any column left contains missing value
merged.isnull().sum()
```

```python
# Unique values that each column contains
merged.nunique()
```
